### Histograma de variables clave

```python
import matplotlib.pyplot as plt
import seaborn as sns

variables_clave = ['age', 'balance', 'duration', 'campaign']
binss = [20,100,100,50]
for i,var in enumerate(variables_clave):
    plt.figure()
    sns.histplot(datos[var],bins=binss[i])
    plt.title(f'Distribución de {var}')
    plt.grid(False)
    plt.show()
```

### Boxplots para variables clave en relación con la contratación 

```python
for var in ['age', 'balance', 'duration']:
    plt.figure()
    sns.boxplot(x='y', y=var, data=datos)
    plt.title(f'Boxplot de {var} por Contratación de Depósito (y)')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
df = pd.read_csv('bank-full.csv', sep=';')

# Variables categóricas
categorical_vars = ['job', 'marital', 'education']

for cat_var in categorical_vars:
    plt.figure(figsize=(10, 6))

    # Contar las ocurrencias de 'yes' y 'no'
    counts = df.groupby([cat_var, 'y']).size().unstack(fill_value=0)

    # Crear el primer eje para las cuentas de 'yes'
    ax1 = plt.gca()  # Obtiene el eje actual
    ax1.bar(counts.index, counts['yes'], color='blue', alpha=0.7, label='Yes', width=0.4)
    ax1.set_ylabel('Cuentas de Yes')
    ax1.set_xlabel(cat_var)
    ax1.set_title(f'Distribución de {cat_var} en relación con Contratación de Depósito (y)')
    
    # Crear un segundo eje para las cuentas de 'no'
    ax2 = ax1.twinx()  # Crea un eje y compartido
    ax2.bar(counts.index, counts['no'], color='orange', alpha=0.7, label='No', width=0.4, align='edge')
    ax2.set_ylabel('Cuentas de No')

    # Ajustar los ticks del eje x
    ax1.set_xticks(range(len(counts.index)))
    ax1.set_xticklabels(counts.index, rotation=45)
    
    # Agregar leyenda
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')

    plt.show()

```

### Efectividad de las campañas por mes

```python
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.pyplot as plt
import seaborn as sns

# Orden de los meses
meses_orden = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

# Filtrar los datos para contar los valores de 'y' como "sí" y "no" por mes
df_counts = df.groupby(['month', 'y']).size().unstack().reindex(meses_orden).fillna(0)
df_counts.columns = ["No", "Sí"]

# Crear la figura y los ejes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Gráfico de barras para "No" en el eje primario
df_counts["No"].plot(kind='bar', color='skyblue', ax=ax1, position=1, width=0.4)
ax1.set_ylabel('Cantidad de No')
ax1.set_xlabel('Mes de Contacto')
ax1.set_title('Cantidad de Contrataciones de Depósitos por Mes')
ax1.legend(["No"], loc="upper left")

# Crear un segundo eje para los "Sí"
ax2 = ax1.twinx()
df_counts["Sí"].plot(kind='bar', color='salmon', ax=ax2, position=0, width=0.4)
ax2.set_ylabel('Cantidad de Sí')
ax2.legend(["Sí"], loc="upper right")

plt.show()


```

### Correlación

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar las columnas de interés para calcular la correlación
cols_interes = ['balance', 'age', 'duration', 'campaign', 'y']
corr = datos[cols_interes].corr()

# Graficar el heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, cmap="Blues", annot=True)
plt.title('Matriz de Correlación')
plt.show()
```
